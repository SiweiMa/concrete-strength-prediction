```python
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport
```

```python
df_chem = pd.read_excel('1-s2.0-S0950061820339581-mmc1.xlsx', 'Table A2')
```

```python
def del_all_missing_col(df):
    cols_w_missing = list(df.columns[df.isnull().all()])
    cols = list(df.columns)

    for col in cols_w_missing:
        cols.remove(col)
    return df[cols]
```

```python
df_chem = del_all_missing_col(df_chem)
```

```python
# set up the hierarchical index of df_chem

df_chem_length = df_chem.shape[0] - 6
df_chem_index_level = range(0, df_chem_length)

# df_chem_column_level_0 = df_chem.iloc[4, :].fillna(method='ffill').to_numpy()

df_chem_column = df_chem.iloc[5, :].fillna(method='ffill').to_numpy()

column_arrays = ['Source', 'Year', 'Type'] + list(df_chem_column)[3:]

df_chem_all_values = df_chem.iloc[6:, :].values

df_chem = pd.DataFrame(df_chem_all_values, index=df_chem_index_level, columns=column_arrays)

df_chem['Year'] = df_chem['Year'].astype(str)

df_chem
```

```python
df_SCC = pd.read_csv('df_SCC_strength.csv')


unique_author_year_in_all_data = df_SCC.iloc[1:, 1:3].drop_duplicates()

df_chem_w_strength = pd.merge(df_chem, unique_author_year_in_all_data, on=('Source', 'Year'))

df_chem_w_strength
```

```python
# fill nan in Specific surface area (m2/kg)
component_names = df_chem_w_strength['Type'].value_counts().index

for component in component_names:
    df_chem_w_strength['Specific surface area (m2/kg)'][df_chem_w_strength['Type']==component] = df_chem_w_strength['Specific surface area (m2/kg)'][df_chem_w_strength['Type']==component]\
                                                                                .fillna(df_chem_w_strength['Specific surface area (m2/kg)'][df_chem_w_strength['Type']==component].mean() )
```

```python
# fill nan in Specific surface area (m2/kg)
component_names = df_chem_w_strength['Type'].value_counts().index

for component in component_names:
    df_chem_w_strength['Specific gravity'][df_chem_w_strength['Type']==component] = df_chem_w_strength['Specific gravity'][df_chem_w_strength['Type']==component]\
                                                                                .fillna(df_chem_w_strength['Specific gravity'][df_chem_w_strength['Type']==component].mean() )
```

```python
# drop Median particle size (μm) column

df_chem_w_strength = df_chem_w_strength.drop('Median particle size (μm)', axis=1)
```

```python
# fill nan with zero
df_chem_w_strength.fillna(0, inplace=True)
df_chem_w_strength
```

```python
df_chem_w_strength.to_csv('df_chem_w_strength.csv')
```
